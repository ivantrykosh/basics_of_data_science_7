### Індексно-орієнтований вибір

```python
# Вибір всіх полів останнього рядка за його індексом
athletics.iloc[-1]
```

```python
# Вибір стовпця 'participant_name' за його індексом
athletics.iloc[:, 11]
```

```python
# Вибір стовпця 'result' за його індексом та одночасний зріз (вибір перших 16 результатів)
athletics.iloc[:16, 16]
```

```python
# Вибір стовпця 'result' за його індексом та одночасний зріз (вибір результатів між 16 та 23)
athletics.iloc[16:24, 16]
```

```python
# Вибір стовпця 'rank' за його індексом та одночасний зріз (вибір результатів між 10 та 100 з кроком 7)
athletics.iloc[10:100:7, 15]
```

```python
# Вибір зрізу рядків (перші 5)
athletics.iloc[:5]
```

```python
# Вибір зрізу рядків (останнні 5)
athletics.iloc[-5:]
```

### Мітко‐орієнтований вибір

```python
# Вибір першого елемента стовпця 'event_name'
athletics.loc[0, 'event_name']
```

```python
# Вибір підмножини стовпців та усіх їх значень
athletics.loc[:, ['event_name','participant_name', 'rank']]
```

```python
# Вибір підмножини стовпців та їх перших 8 значень
athletics.loc[:7, ['event_name','participant_name', 'rank']]
```

```python
# Вибір підмножини стовпців та їх значень від 0 до 16 з кроком 8
athletics.loc[:16:8, ['event_stage','participant_name', 'rank']]
```

### Маніпулювання індексом

```python
# Створення індексу date у датафреймі
athletics.set_index("date")
```

### Умовний вибір

```python
# Формування піднабору даних, де participant_country - Україна
athletics.loc[athletics.participant_country == "Ukraine"]
```

```python
# Формування піднабору даних, де participant_country - Україна та stage - фінал
athletics.loc[(athletics.participant_country == "Ukraine") & (athletics.stage == "Final")]
```

```python
# Формування піднабору даних, де event_name - жіночі стрибки у висоту або стать - змішана
athletics.loc[(athletics.event_name == "Women's High Jump") | (athletics.gender == "X")]
```

```python
# Усі результати Ярослави Магучіх
athletics.loc[athletics.participant_name.isin(['MAHUCHIKH Yaroslava'])]
```

```python
# Усі результати чоловічих та жіночих стрибків у висоту
athletics.loc[athletics.event_name.isin(["Women's High Jump", "Men's High Jump"])]
```

```python
# Вибір результатів суто чоловіків або жінок (немає змішаних дисциплін)
athletics.loc[~athletics.gender.isin(["X"])]
```

```python
# Вибір усіх рядків, де result - null
athletics.loc[athletics.result.isnull()]
```

```python
# Вибір усіх рядків, де rank - не null
athletics.loc[athletics['rank'].notnull()]
```
