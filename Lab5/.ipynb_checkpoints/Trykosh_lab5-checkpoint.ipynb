```python
# Імпортуємо бібліотеки
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Етап підготовки даних

```python
# Імпортуємо набір даних: https://www.kaggle.com/datasets/fernandol/countries-of-the-world
countries_set = pd.read_csv('countries of the world.csv')
```

```python
# Початкові дані. Бачимо, що потрібні нам стопвці Birthrate та Deathrate треба привести до потрібного типу
countries_set.head()
```

```python
# Замінимо коми на крапки, щоб у подальному привести значення до типу float
countries_set['Birthrate'] = countries_set['Birthrate'].str.replace(',', '.').astype(float)
countries_set['Deathrate'] = countries_set['Deathrate'].str.replace(',', '.').astype(float)
```

```python
# Початкові дані
countries_set.head()
```

```python
# Прикінцеві дані
countries_set.tail()
```

```python
# Перегляд розміру набору даних
countries_set.shape
```

```python
# Перевірка повноти набору даних
# З отриманих результатів бачимо, що певні рядки містять нульові дані, тому треба їх видалити
countries_set.info()
```

```python
# Видалимо усі рядки, у яких є нульові значення
countries_set.dropna(inplace=True)
```

```python
# Перевірка повноти набору даних
# З отриманих результатів бачимо, що всі 179 рядків містять ненульові дані
countries_set.info()
```

```python
# Зробимо ще одну перевірку, яка підраховує кількість NaN значень по стовпцях. Бачимо, що немає NaN значень
countries_set.isnull().sum()
```

```python
# Зробимо підмножину даних з 15 (Birthrate) та 16 (Deathrate) стовпців основного набору, які використовуватимемо, як ознаки
x = countries_set.iloc[:, [15, 16]].values
x
```

### Етап побудови моделі

```python
# Імпортуємо KMeans та створимо список кластерів з їх інерцією
from sklearn.cluster import KMeans
clalist = []
```

Щоб визначити оптимальну кількість кластерів, застосуємо метод Elbow

```python
# inertia_ - це метод, що вимірює інерцію кластера (показник, наскільки близько точки даних згруповані навколо своїх центрів кластерів)
for i in range(1, 11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(x)
    clalist.append(kmeans.inertia_)    
```

```python
# Побудуємо графік з кількістю кластерів та їх інерціями
# За графіком бачимо, що оптимальна кількість кластерів - 5, бо після 5 кластерів інерція зменшується повільно
plt.plot(range(1, 11), clalist)
plt.title('Метод Elbow')
plt.xlabel('Кількість кластерів')
plt.ylabel('clalist')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters=5, init='k-means++', random_state=0)
y_kmeans = kmeansmodel.fit_predict(x)
```

### Етап візуалізації

```python
# Візуалізуємо результати кластеризації
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y_kmeans == 4, 0], x[y_kmeans == 4, 1], s = 100, c = 'black', label = 'Cluster 5')
plt.scatter(kmeansmodel.cluster_centers_[:, 0], kmeansmodel.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Кластери країн')
plt.xlabel('Рівень народжуваності')
plt.ylabel('Рівень смертності')
plt.legend()
plt.show()
```
