```python
# Застосування відображення "map" до стовпця "result_type" (приведення типу результату до нижнього регістру)
athletics['result_type'].map(lambda p: p.lower())
```

```python
# Застосування відображення "apply" до датафрейму (обмін значеннями країни та коду країни у кожному рядку)
def swap_values(row):
    row['participant_country_code'], row['participant_country'] = row['participant_country'], row['participant_country_code']
    return row

athletics.apply(swap_values, axis='columns')
```

```python
# Вивід перших 5-ти рядків датафрейму
athletics.head()
```

```python
# Віднімання від кожної початкової позиції середнього значення
athletics_start_order_mean = athletics['start_order'].mean()
athletics['start_order'] - athletics_start_order_mean
```

```python
# Вивід змагання, коду країни та результату
athletics['event_name'] + " - " + athletics['participant_country_code'] + " - " + athletics['rank'].astype(str)
```
