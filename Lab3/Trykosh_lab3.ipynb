```python
# Завантажуємо початкові модулі
import pandas as pd
import numpy as np

# Читаємо весь набір даних
athletics = pd.read_csv("archive/results/Athletics.csv")

# Встановлюємо зерно для генератора випадкових чисел
np.random.seed(42) 
```

###  Перше ознайомлення з даними

```python
# Переглянемо перші 5 рядків датасету. Бачимо пропущені дані
athletics.head(5)
```

### Визначення кількості відсутніх даних

```python
# Отримуємо кількість відсутніх елементів даних по стовпцях
missing_values_count = athletics.isnull().sum()

# Виводимо кількість відсутніх елементів у стовпцях
missing_values_count
```

```python
# Визначаємо загальну кількість відсутніх елементів даних
total_cells = np.product(athletics.shape)
total_missing = missing_values_count.sum()

# Визначаємо відсоток відсутніх елементів даних
percent_missing = (total_missing / total_cells) * 100
percent_missing
```

### Визначення причини відсутності даних

```python
# Усі інші стовпці, крім venue, мають пропущені дані за визначенням (їх просто не існує)
missing_values_count.loc['venue']
```

### Видалення відсутніх елементів даних

```python
# Видаляємо всі рядки, що містять відсутні значення
athletics.dropna()
```

```python
# Видаляємо всі стовпці, що містять відсутні значення
columns_with_na_dropped = athletics.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# Визначимо кількість втрачених даних
print("Кількість стовпців : %d \n" % athletics.shape[1])
print("Кількість видалених стовпців: %d" % columns_with_na_dropped.shape[1])
```

```python
# Заповнення відсутніх елементів даних нулями
athletics.fillna(0)
```

```python
# Створення піднабору даних
subset_athletics = athletics.loc[athletics.venue.isnull()][['event_name', 'venue']]
subset_athletics.head()
```

```python
# Заповнення стовпця venue правильним значенням
subset_athletics.fillna('Stade de France').head()
```

```python
# Заповнення наступним по черзі значенням
athletics.fillna(method='bfill', axis=0).fillna(0)
```
